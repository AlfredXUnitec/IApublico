```python
class Point3D(object):
    def __init__(self,a,b,c):
        self.x = a
        self.y = b
        self.z = c
    def __repr__(self):
        return "Point3D(%d, %d, %d)" % (self.x, self.y, self.z)
    def __str__(self):
        return "(%d, %d, %d)" % (self.x, self.y, self.z)
my_point = Point3D(1, 2, 3)
print(repr(my_point)) # __repr__ gets called automatically
print(str(my_point)) # __str__ gets called automatically
```

# Composition. 
In compostion one of the classes is composed of one or more instance of other classes.
In other words one class is container and other class is content and if you delete the container object 
then all of its contents objects are also deleted.
Now lets see an example of composition in Python 3.5. Class Employee is container and class Salary is content.
class Salary:

```python
class Salary:
    def __init__(self,pay):
        self.pay=pay

    def get_total(self):
       return (self.pay*12)

class Employee:
    def __init__(self,pay,bonus):
        self.pay=pay
        self.bonus=bonus
        self.obj_salary=Salary(self.pay)

    def annual_salary(self):
        return "Total: "  +  str(self.obj_salary.get_total()+self.bonus)


obj_emp=Employee(100,10)
print (obj_emp.annual_salary())
```

# Aggregation

Aggregation is a weak form of compostion. If you delete the container object contents objects can live without container object.

Now lets see an example of aggregation in Python 3.5. Again Class Employee is container and class Salary is content.

```python
class Salary:
    def __init__(self,pay):
        self.pay=pay

    def get_total(self):
       return (self.pay*12)

class Employee:
    def __init__(self,pay,bonus):
        self.pay=pay
        self.bonus=bonus

    def annual_salary(self):
        return "Total: "  +  str(self.pay.get_total()+self.bonus)


obj_sal=Salary(100)
obj_emp=Employee(obj_sal,10)
print (obj_emp.annual_salary())
```

# Ejemplo Clases

```python
import datetime # Se usa para objetos de fecha

class Persona:

    def __init__(self, nombre, apellido, fechanac, direccion, telefono, email):
        self.nombre = nombre
        self.apellido = apellido
        self.fechanac = fechanac

        self.direccion = direccion
        self.telefono = telefono
        self.email = email

    def edad(self):

        if hasattr(self, "_edad"):
            return self._edad
        
        hoy = datetime.date.today()
        edad = hoy.year - self.fechanac.year

        if hoy < datetime.date(hoy.year, self.fechanac.month, self.fechanac.day):
            edad -= 1

        return edad

persona = Persona(
    "Jose",
    "Blancas",
    datetime.date(1992, 3, 12), # ao, mes, day
    "No. 12 Short Street, Greenville",
    "555 456 0987",
    "jlblancas@gmail.com"
)

print(persona.nombre)
print(persona.email)
print(persona.edad())
```

# Atributos

Todos los atributos que se definen en una instancia de Persona  son atributos de instancia  se agregan a la instancia cuando el mtodo  init se ejecuta. 

Sin embargo, podemos tambin definir los atributos que se establecen en la Clase. Todos los ejemplares de esa clase compartirn estos atributos. De muchas maneras se comportan igual que los atributos de la instancia, pero hay algunas advertencias que usted debe tener en cuenta.

Definimos atributos de clase en el cuerpo de una clase, en el mismo nivel de sangra que las definiciones de mtodo (un nivel hacia arriba desde el interior de los mtodos):


```python
class Persona:

    TITULOS = ('Dr', 'Sr', 'Sra', 'Srita')

    def __init__(self, titulo, nombre, apellido):
        if titulo not in self.TITULOS:
            raise ValueError("%s no es un titulo valido." % titulo)

        self.titulo = titulo
        self.nombre = nombre
        self.apellido = apellido
```

```python
# Pero podemos accesarlos desde la clase
Persona.TITULOS

# No Podemos accesar los atributos de la clase desde una instancia.

persona.TITULOS

# Notar que un objeto de la clase no tiene acceso a ninguno de los atributos de la instancia.
# Estos son creados cuando la instancia es creada.
# Esto nos dar un error.

Persona.nombre
Persona.apellido
```
